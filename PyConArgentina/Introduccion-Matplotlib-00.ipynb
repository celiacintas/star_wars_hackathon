### Matriz de confusión

La matriz de confusión es una herramienta que permite la visualización del desempeño de un algoritmo. Cada fila de la matriz representa el número de predicciones de cada clase, mientras que cada columna representa a las instancias en la clase real. Uno de los beneficios de las matrices de confusión es que facilitan ver si el sistema se está confundiendo entre clases. [Mas de confusion matrix](https://en.wikipedia.org/wiki/Confusion_matrix)

```python
my_conf_matrix = np.array([[337, 10, 4], [ 4, 401, 20], [ 2, 2, 369]]) 
plt.tick_params(labelsize=20)

ax = plt.imshow(my_conf_matrix, interpolation='nearest', cmap=plt.cm.Reds)
cb = plt.colorbar()
cb.ax.tick_params(axis='both', labelsize=20)
plt.ylabel("True label", fontsize=20)
plt.xlabel("Predicted label", fontsize=20)
plt.xticks(range(3),range(3))
plt.yticks(range(3),range(3))
```
