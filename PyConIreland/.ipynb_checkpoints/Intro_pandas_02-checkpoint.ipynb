```python
nonduplicate_data = data.drop_duplicates()
nonduplicate_data
```

```python
interpolate_data = nonduplicate_data.interpolate(method='nearest', axis=0)
interpolate_data
```

```python
interpolate_data["Class_num"] = pd.factorize(interpolate_data["Class"])[0]
```
